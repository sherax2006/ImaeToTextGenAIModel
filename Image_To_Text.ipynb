```python
# Install necessary libraries
!pip install transformers torch pytesseract opencv-python-headless easyocr
!apt-get install -y tesseract-ocr

from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
import pytesseract
import cv2
import numpy as np
import easyocr
import re
from google.colab import files

# Upload an image file from your local system
uploaded = files.upload()

# Load the uploaded image
for filename in uploaded.keys():
    image = Image.open(filename)
    break  # Assuming only one image is uploaded

# Convert PIL image to OpenCV format
image_cv = np.array(image)

# Convert image to grayscale
gray_image = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur to reduce noise
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Apply adaptive thresholding
binary_image = cv2.adaptiveThreshold(
    blurred_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
)

# Convert back to PIL image
processed_image = Image.fromarray(binary_image)

# Load the pre-trained TrOCR model and processor
model_name = "microsoft/trocr-large-printed"
processor = TrOCRProcessor.from_pretrained(model_name)
model = VisionEncoderDecoderModel.from_pretrained(model_name)

# Process the image with TrOCR
pixel_values = processor(images=image, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)
text_trocr = processor.batch_decode(generated_ids, skip_special_tokens=True)
print("TrOCR text:", text_trocr[0])

# Use Tesseract OCR on the processed image
text_tesseract = pytesseract.image_to_string(processed_image)
cleaned_text_tesseract = re.sub(r'\s+', ' ', text_tesseract).strip()
print("Processed Tesseract OCR text:", cleaned_text_tesseract)

# Use EasyOCR to extract text from the image
reader = easyocr.Reader(['en'], gpu=True)
result = reader.readtext(image_cv)
easyocr_text = ' '.join([detection[1] for detection in result])
print("EasyOCR text:", easyocr_text)
```
